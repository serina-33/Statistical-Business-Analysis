```python
import pandas as pd
import numpy as np

# Creating a sample dataset
np.random.seed(42)

data = pd.DataFrame({
    "Week": range(1, 101),
    "Marketing_Spend": np.random.randint(5000, 30000, 100),
    "Revenue": np.random.randint(40000, 80000, 100)
})

# Simulating sales before and after a marketing campaign
# The length of the arrays should match the number of rows in the DataFrame (100).
data["Sales_Before"] = np.random.randint(45000, 55000, 100)
data["Sales_After"] = np.random.randint(50000, 65000, 100)

data.head()
```

```python
# Descriptive statistics
mean_sales = data["Revenue"].mean()
median_sales = data["Revenue"].median()
mode_sales = data["Revenue"].mode()[0]
std_sales = data["Revenue"].std()

mean_sales, median_sales, mode_sales, std_sales
```

```python
import matplotlib.pyplot as plt

# Histogram
plt.hist(data["Revenue"])
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

```python
correlation = data["Marketing_Spend"].corr(data["Revenue"])
correlation
```

```python
from scipy.stats import ttest_ind

# Running a t-test
t_stat, p_value = ttest_ind(data["Sales_Before"], data["Sales_After"])

t_stat, p_value
```

```python
import numpy as np

sales = data["Revenue"]

mean = sales.mean()
std = sales.std()
n = len(sales)

# 95% confidence interval formula
ci_lower = mean - 1.96 * (std / np.sqrt(n))
ci_upper = mean + 1.96 * (std / np.sqrt(n))

ci_lower, ci_upper
```
